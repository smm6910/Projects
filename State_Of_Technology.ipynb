## Since I'm entering the tech field, I wanted to explore how other data scientists and engineers feel about the industry. Resulting from a 2017 survey created by Kaggle, this dataset consists of a series of multiple choice and free response questions regarding happiness levels, work challenges and general recommendations for the most useful burgeoning machine learning tools. With 16,000 respondents in over 25 countries, a lot of data was available. The "free form" was the free response portion of the survey -- subjects could respond however they wanted. 



#### The fine print: 

While I was able to glean a lot of information, the data did not come without its limitations. 

Kaggle did not require people to respond to every question, whether multiple choice or free response; consequently, a lot of data was missing. A brief glance over the responses show many people either stopped at a particular question or skipped around, answering questions they liked. Though I was unable to confirm this, I also wonder whether there was a language barrier, given many subjects were from non-English speaking countries. 

There was a strong voluntary response bias. Most respondents were males from the United States and India, which would not accurately represent tech workers worldwide. Furthermore, because this survey was available to anyone with an interest in technology, many respondents had never worked in the field, likely decreasing the accuracy of results as well. 





#### If I had conducted this survey....

There are a few things I would have changed. I would have taken a random sampling of male and female data scientists and engineers around the world, to more accurately represent the realities, and people's opinions,of the technology field. 

I would also require every question be answered, for more accurate results, and would only include respondents who were currently working in the field.
